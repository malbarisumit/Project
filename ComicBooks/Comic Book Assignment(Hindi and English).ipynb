Read the caption from excel

```python
import pandas as pd
data = pd.read_excel('comicBooks/Assignment/ComicDialogue.xls')
data.head(10)
```

```python
Create and save Captiopned images in a different folder
```

```python
from PIL import Image as pili, ImageDraw as pild, ImageFont as pilf 
from PIL import Image, ImageFont, ImageDraw

mypath = 'comicBooks/Assignment/borderImages/'
mynewpath = 'comicBooks/Assignment/borderCaptionImages/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
for i in range(0,102):
    caption = data.values[i][0]

    TINT_COLOR = (0, 0, 0)  # Black
    TRANSPARENCY = .25  # Degree of transparency, 0-100%
    OPACITY = int(255 * TRANSPARENCY)

    

    img = pili.open(mypath+onlyfiles[i]).convert('RGBA')

    overlay = pili.new('RGBA', img.size, TINT_COLOR+(0,))
    draw = pild.Draw(overlay)
    font = pilf.truetype("arial.ttf", 25)
    #x, y = (img.width - 510, img.height-100)
    text = caption
    
    w, h = font.getsize(text)
    num_lines = len(text.split('\n'))
    print(num_lines, w, h)
    x, y = 70, img.height - (num_lines-0.1*num_lines)*h -45
    #draw.rectangle((x, y, x + w, y + h), fill='black')
    #draw.rectangle((x, y, x + w, y + 4*h), fill=TINT_COLOR+(OPACITY,))
    draw.rectangle((x, y, x + img.width - 200, y + (num_lines-0.1*num_lines)*h), fill=TINT_COLOR+(OPACITY,))
   # draw.text((x, y), text.split('\n')[0]+"\n"+text.split('\n')[1], fill=(209, 239, 8), font=font)
    #draw.text((x, y), text, fill=(209, 239, 8), font=font)
    draw.text((x, y), text, fill=(209, 239, 8), font=font)
    # Alpha composite these two images together to obtain the desired result.
    img = pili.alpha_composite(img, overlay)
    img = img.convert("RGB") # Remove alpha for saving in jpg format.
    onlyfiles[i] = onlyfiles[i][:1] + 'c' + onlyfiles[i][1:]
    img.save(mynewpath+onlyfiles[i])
```

Create and Save Horizontal and vertical stacks of images

```python
from os import listdir
from os.path import isfile, join

#mypath = 'comicBooks/bimage/'
#onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles = lt
vlist=[]
x=0
y=0
while 0 < len(onlyfiles):
    lines = 0
    print("new page!")
    num_cols = old_num_cols = 0
    while lines < 6 and 0 < len(onlyfiles):
        lines += 1
        while num_cols == old_num_cols:
            num_cols = np.random.randint(2, 5)
        old_num_cols = num_cols
        window = onlyfiles[:num_cols]
        imgs    = [ PIL.Image.open(i) for i in window ]

# pick the image which is the smallest, and resize the others to match it (can be arbitrary image shape here)
        min_shape = sorted( [(np.sum(i.size), i.size ) for i in imgs])[0][1]
        imgs_comb = np.hstack( (np.asarray( i.resize(min_shape) ) for i in imgs ) )
        imgs_comb = PIL.Image.fromarray( imgs_comb)
        imgs_comb.save( 'comicBooks/Assignment/horizontalStackImages/'+str(x)+'.png' ) 
        vlist.append('comicBooks/Assignment/horizontalStackImages/'+str(x)+'.png' )
        x += 1
# save that beautiful picture
        #imgs_comb = PIL.Image.fromarray( imgs_comb) 
        onlyfiles = onlyfiles[num_cols:]
    imgs1    = [ PIL.Image.open(j) for j in vlist ]

# pick the image which is the smallest, and resize the others to match it (can be arbitrary image shape here)
    min_shape = sorted( [(np.sum(j.size), j.size ) for j in imgs1])[0][1]
    imgs_comb1 = np.vstack( (np.asarray( j.resize(min_shape) ) for j in imgs1 ) )
    imgs_comb1 = PIL.Image.fromarray( imgs_comb1)
    imgs_comb1.save( 'comicBooks/Assignment/verticalImages/page'+str(y)+'.png' )
    vlist = []
    y += 1
```

```python
Convert images to pdf
```

```python
from PIL import Image
from fpdf import FPDF

#pdf = FPDF()
cover = Image.open('comicBooks/Assignment/verticalImages/page0.png')
width, height = cover.size
pdf = FPDF(unit = "pt", format = [width, height])
    
# imagelist is the list with all image filenames
imagelist = ['comicBooks/Assignment/verticalImages/page0.png','comicBooks/Assignment/verticalImages/page1.png',
            'comicBooks/Assignment/verticalImages/page2.png','comicBooks/Assignment/verticalImages/page3.png',
            'comicBooks/Assignment/verticalImages/page4.png','comicBooks/Assignment/verticalImages/page5.png',]

for image in imagelist:
    pdf.add_page()
    pdf.image(image, 0, 0, width, height)
pdf.output("comicBooks/Assignment/ComicBook/Robo.pdf", "F")
print("done!")
print(width, height)
```
