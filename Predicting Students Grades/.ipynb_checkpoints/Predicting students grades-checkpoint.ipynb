<font size= 3>If the correlation coefficient is equal to 0, there is no linear correlation between the two variables (if one variable changes value, that doesn't give us any information about if the other variable is going to change as well or not). <font size= 3>Also being in romantic relation and gender should not affect your grades, hence <b>sex and romantic</b> is removed. Also going for higer education should not affect grades because there are people who wants to go for higher education who will need grades but there are also students who wants to join a company which also needs grades. Also guardian can be anyone so, it should not have any effect on the students grade. <b>Reason</b> for joining should also be removed as someone with higher grades who wants to join a college to a near by uni/school and some with lower grade can also have same mentality</font>

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

<font size = 5><b>6) Create training and testing sets, with an 80-20 split</b></font>

```python
# Need to remove NA
# Regression - Try to predict Mean_Humidity (y) from non humidity columns (X)
# Get training set (X_train)
def valid(col):
    return col != 'G3'
midterm_dummy = midterm_dummy.dropna()

# we remove the last observati; L
on (last row) because there is no tomorrow for the last row to predict!
X = midterm_dummy[[x for x in midterm_dummy.columns if valid(x)]]

# for the dependent variable, we remove the NA, so that removes the last row, too
y = midterm_dummy.G3.dropna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=243)
```

```python
X.T
```

```python
y
```

<font size = 5><b>7) Build a random forest model to predict student grades</b></font>

```python
# Create a model 
rf_model = RandomForestRegressor()

# Train the model
rf_model.fit(X_train, y_train)
```

```python
# Get R2 measure (indicator of accuracy 1 is perfect, 0 is horrible)
rf_model.score(X_test, y_test)
```

```python
y_pred = rf_model.predict(X_test)
```
