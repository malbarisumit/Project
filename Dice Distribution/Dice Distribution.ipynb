Uniform Distribution

```python
x = [1,2,3,4,5,6]
counts, bins = np.histogram(x, bins= 50)
bins = bins[:-1] + (bins[1] - bins[0])/2
probs = counts/float(counts.sum())
print(probs.sum()) # 1.0
plt.bar(bins, probs)
plt.show()
```

```python
probs = np.full((6), 1/6) #Return a new array of shape (6) and type, filled with 1/6
face = [1,2,3,4,5,6]
plt.bar(face, probs)
plt.ylabel('Probability', fontsize=12)
plt.xlabel('Die Roll Outcome', fontsize=12)
plt.title('Fair Die follows a Uniform Distribution', fontsize=12)
axes = plt.gca()
axes.set_ylim([0,1])
```

Bernouli distribution for getting a face(eg: face 4) (p(4)=1/6  and rest probabilty=5/6). Failure =0, success =1

```python
bern = stats.bernoulli.rvs(size=10000, p=1/6)
ax = sns.distplot(bern, kde=False)
ax.set(xlabel='Bernoulli Distribution', ylabel = 'Frequency')
```

Binomial Distribution

```python
binom_data = stats.binom.rvs(size=10000, n=10, p=1/6, random_state=0)
pd.Series(binom_data).plot(kind = "hist", bins=50)
```

Normal or Guassian Distribution

```python
from pymc3 import Normal
```

```python
roll = pd.Series([1,2,3,4,5,6])
mean = roll.mean()
std = roll.std()
print(mean)
print(std)
```

```python
from pymc3 import Normal

y = Normal.dist(mu=-std, sd=mean)
samples = y.random(size=10000)
print(samples.mean())
print(samples.std())
plt.plot(samples);
```

```python
plt.hist(samples);
```

```python
normal_dist = stats.norm.rvs(size=10000, loc=mean, scale=1, 
                             random_state=0) 
pd.Series(normal_dist).plot(kind="hist", bins=100, 
                                label="Mean = {:f}".format(mean))
plt.xlabel('Distribution', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title("Normal Distribution varying the mean")
plt.legend()
```

Poisson distribution

```python
from pymc3 import Poisson

x = Poisson.dist(mu=mean)
samples = x.random(size=100)
print(samples.mean())
print(samples.std())
plt.plot(samples)
```

2) Weighted Die

```python
probs = np.array([0.15, 0.125, 0.10, 0.30, 0.250, 0.075])
face = [1,2,3,4,5,6]
plt.bar(face, probs)
plt.title('Loaded/ Weighted die (not uniform distribution)', fontsize=12)
plt.ylabel('probability', fontsize=12)
plt.xlabel('Die Roll Outcome', fontsize=12)
axes = plt.gca()
axes.set_ylim([0,1])
```

```python
n = np.arange(1, 7)
mean = 3.5
gaussian_dice = stats.norm.pdf(n, mean)
plt.plot(n, gaussian_dice)

face = [1,2,3,4,5,6]
plt.bar(face, gaussian_dice)
plt.title('Loaded dice (gausian distribution)', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xlabel('Dice Roll Outcome', fontsize=12)
axes = plt.gca()
axes.set_ylim([0,1])
```
